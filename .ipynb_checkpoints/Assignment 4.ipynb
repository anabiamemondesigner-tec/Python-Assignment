# Q4.
You are managing a list of students enrolled in two different courses. You need to perform various set operations to understand the student enrollment. Task: Create two sets: one for students in "Course A" and one for students in "Course B". Find the students who are in "Course A" but not in "Course B" (difference). Find students who are only in one of the two courses (symmetric difference). course_a = {"John", "Alice", "Bob", "David"} course_b = {"Alice", "Eve", "Charlie", "David"}

Using set methods

• Find students who are in Course A but not in Course B (difference)

• Find students who are only in one of the two courses (symmetric difference)

Expected Output:

Students in Course A but not in Course B: {'Bob', 'John'}

Students only in one of the two courses: {'Eve', 'John', 'Bob', 'Charlie'}

```python
course_a = {"John", "Alice", "Bob", "David"}
```

```python
course_b = {"Alice", "Eve", "Charlie", "David"}
```

```python
Student = course_a.difference(course_b)
```

```python
print("Students in Course A but not in Course B:", Student)
```

```python
Students = course_b.symmetric_difference(course_a)
```

```python
print("Students only in one of the two courses:", Students)
```

# Q10. Set Comparison
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7}

Find common elements.
Find elements only in Set A.
Find elements only in Set B.

```python
A = {1, 2, 3, 4, 5}
```

```python
B = {4, 5, 6, 7}
```

```python
A_diff = A.difference(B)
```

```python
print("Numbers in A variable but not in B Variable:", A_diff)
```

```python
B_diff = B.difference(A)
```

```python
print("Numbers in B variable but not in A Variable:", B_diff)
```

```python
common_elements = A.intersection(B)
```

```python
print("Numbers is common in A and B Variable:", common_elements)
```
